# Bar Chart In Matplotlib

```python
import numpy as np
```

```python
company = ['Google','Amazon','Oracle','TCS']
revenue = [90,136,89,27]
profit = [48,2,34,12]
```

```python
plt.ylabel('revenue(in $)')
plt.title('US Tech Stacks')
plt.bar(company,revenue,label='Revenue',color = 'green') #default color is blue
plt.bar(company,profit,label='Profit',color='red')
plt.legend()
#it is plotting in the same chart
```

If we do this without using numpy module we will get in same chart.so inorder to draw side by side we do following


```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(in $)')
plt.title('US Tech Stacks')
plt.bar(xpos-0.2,revenue,width = 0.4,label='Revenue',color = 'green') #default color is blue
plt.bar(xpos+0.2,profit,width = 0.4,label='Profit',color='red')
plt.legend()
```

# Horizontal Bar charts

```python
plt.yticks(xpos,company)
plt.xlabel("Revenue(in $)")
plt.title("US TECH STOCK")
plt.barh(xpos-0.2,revenue,height=0.4,label="Revenue")
plt.barh(xpos+0.2,profit,height=0.4,label="Profit")
plt.legend(shadow=True)
```

# Histogram Charts

These are bar charts where y axis represnts the frequency of a variable.here we have a concept of bins,buckets

```python
blood_sugar_man = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_woman = [67,98,89,120,133,150,84,69,89,79,120,112,100]
#plt.hist(blood_sugar)
#plt.hist(blood_sugar,bins=3,rwidth =0.95)
plt.xlabel('Sugar_Level')
plt.ylabel('Total No Of Patients ')
plt.title('Blood Sugar Levels')
plt.hist([blood_sugar_man,blood_sugar_woman],bins=[80,100,125,150],rwidth =0.95,color = ['green','orange'],label = ['Men','Women'])
plt.legend()
#default bins =10 and we can also reduce
#rwidth property  is relatively width  of a bar comapred to bin size. its maximum value is 1.
#histtype = 'step'
#Orientation property says how histogram chart should look like i.e Horizontal or vertical.default is vertical.
#if we want orientation in horizontal keep orientation = 'horizontal'.
```

# Pie Chart

```python
exp_value = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet Bill','Car','Other Utilities']
plt.pie(exp_value,labels = exp_labels)
plt.axis('equal') #inorder to look nice in a circle.
plt.show() #inorder to get-rid of matter.only we will get pie chart here.
#radius parameter used to increase the size of pie chart .default value is 1
```

```python
plt.axis('equal')
plt.pie(exp_value,labels=exp_labels,radius=1.5,shadow = True,autopct = '%1.1f%%',explode = [0,0,0,0,0])
plt.show() 
#autopct = '%f%%' it is for getting the value of that space.
#if you want integer just put autopct = '%0.0f%%'
```

```python
plt.axis('equal')
plt.pie(exp_value,labels=exp_labels,radius=1.5,shadow = True,autopct = '%1.1f%%',explode = [0,0.2,0,0.3,0])
plt.show() 
#explode parameter used to exploded something from chart like explode = [0,0,0,0,0]-> it wont effect and it is default..i.e(no.of labels) in explde parameter
```

```python
plt.axis('equal')
plt.pie(exp_value,labels=exp_labels,radius=1.5,shadow = True,autopct = '%1.1f%%',explode = [0,0.2,0,0.3,0],startangle = 45)
plt.show() 
#startangle tells from which angle to start the chart.
```

# Save Chart To a File Using savefig

```python
plt.axis('equal')
plt.pie(exp_value,labels=exp_labels,radius=1.5,shadow = True,autopct = '%1.1f%%',explode = [0,0,0,0,0])
plt.savefig('E:/piechart.png',bbox_inches='tight',pad_inches=2,Transparent = None)
#bbox_inches is for to set the chart in borsers in the saved file and pad_inches for padding& Transparent for background color.
```

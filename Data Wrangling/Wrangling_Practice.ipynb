# DATA WRANGLING IN PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

```python
df= pd.read_csv(filename)
df.head
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style",
                      "drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type",
                      "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
                      "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df= pd.read_csv(filename,names= headers)
```

```python
df.head()
```

```python
df.replace("?",np.nan,inplace= True) #we have to mention inplace if not it wont change in actual memory.
df.head(5)
```

```python
#checking wherever the null value exists in the dataframe.
missing_values =  df.isnull()
missing_values.head(5)
#True means there null value exists.
```

```python
for column in df.columns.values.tolist():
    print(column)
    print(df[column].value_counts())
    print("")
    
#we are looping over each column and take each value in that column and it displaces count.
```

```python
check_missing = df.isnull()
for column in check_missing.columns.values.tolist():
    print(column)
    print(check_missing[column].value_counts())
    
#checking in each column how many not nulls & Nulls are  present.
```

```python
#mean for specific column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("average of normalized-losses  :", avg_norm_loss)
```

```python
#if we want to replace null values with average of that column.
df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace= True)
```

```python
avg_bore = df["bore"].astype("float").mean(axis=0)
print("average of bore  :", avg_bore)
```

```python
#if we want to replace null values with average of that column.
df["bore"].replace(np.nan,avg_bore,inplace= True)
```

```python
avg_horsepower = df["horsepower"].astype("float").mean(axis=0)
print("average of horsepower  :", avg_horsepower)
```

```python
#if we want to replace null values with average of that column.
df["horsepower"].replace(np.nan,avg_horsepower,inplace= True)
```

```python
#checking how many are present for 2 doors and 4 doors in num-of-doors column
df["num-of-doors"].value_counts()
```

```python
#what is the maximum values found in that column
df["num-of-doors"].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.nan,"four",inplace= True)
```

```python
#simply drop whole row with NaN in price column
df.dropna(subset=["price"],axis=0,inplace=True)
df.head()
```

```python
#checking the datatypes for all columns
df.dtypes
```

```python
#We can also change the datatype
df[["bore","stroke"]] = df[["bore","stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
```

```python
df.dtypes
```

```python
#Converting /Tranforming the data to other type.
#Convert mpg to L/100km by mathematicala expression(235 divided by mpg)
df["city-L/100km"] = 235/df["city-mpg"]
df.head()
#check transformed data.a new column is created..
```

```python
#Normalizing the value by keeping the same properties.
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
```

# Binning

Here,we convert numerical/integer values to categorical values in that column

```python
df["horsepower"] = df["horsepower"].astype(int,copy=True)
#creating bins
bins = np.linspace(min(df["horsepower"]),max(df["horsepower"]),4)
group_names = ['Low','Medium','High']
#creating a new column
df["horsepower-binned"]= pd.cut(df["horsepower"],bins,labels= group_names,include_lowest = True)
df[["horsepower","horsepower-binned"]].head(20)
```

```python
df["horsepower"].value_counts()
```

```python
df["horsepower-binned"].value_counts()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(group_names,df["horsepower-binned"].value_counts())
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower-binning")
plt.legend()
```

```python
plt.hist(df["horsepower-binned"],bins=3)
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower-binning")
plt.legend()
```

# Dummies


It is reverse of binning.converting categorical values to integer/numeric values.

```python
dummy_1 = pd.get_dummies(df["fuel-type"])
dummy_1.head()
```

```python
dummy_1.rename(columns={'fuel-type-diesel' :'gas','fuel-type-diesel' :'diesel'},inplace= True)
dummy_1.head()
```

```python
#Merge dataframe 'df' and dummy_1
df= pd.concat([df,dummy_1],axis=1)
#drop original column fuel-type from "df"
df.drop("fuel-type",axis=1,inplace=True)
```
